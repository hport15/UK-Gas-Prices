```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
filename = '/content/drive/MyDrive/Data Science/Data/Weekly Fuel Prices.xlsx'
df = pd.read_excel(filename)
df
```

```python
df.info()
```

The missing rows were irrelevant so dropped.

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
from numpy.ma.core import size
plt.figure(figsize=(20,10))
plt.plot(df['Date'],df['Petrol (USD)'],marker='s', markersize=3, c='c', label='Gas')
plt.plot(df['Date'],df['Diesel (USD)'],marker='o',c='k',markersize=3, label='Diesel')
plt.xlabel('Date',fontsize=14)
plt.ylabel('Prices in USD',fontsize=14)
plt.title('UK Gas Prices Over Time',fontsize=16)
plt.grid(c='b',alpha=.2)
plt.tight_layout()
plt.legend()
plt.savefig('UKgasdiesel.png', dpi = 300);
```
